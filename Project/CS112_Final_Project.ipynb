```python
import cProfile
def FindSubsetSum(set, n, sum):
	subset =([[(False, []) for i in range(sum + 1)] for i in range(n + 1)])
	for i in range(n + 1): subset[i][0] = (True, [])
	for i in range(1, n + 1):
		for j in range(1, sum + 1):
			if (j < set[i - 1]) or (subset[i - 1][j][0]):
				subset[i][j] = subset[i - 1][j]
			elif subset[i - 1][j - set[i - 1]][0]:
				#traceback 
				subset[i][j] = (True, [i - 1] + subset[i - 1][j - set[i - 1]][1])
	return subset[n][sum]


public_key = [10042, 2904, 1271, 2542]
answer = 4 # 0 1 0 0
C = 2904
STEP = 100
CORRECTED_CASES = 0
NUMBER_OF_CASES = 40
extended = [0 for x in range(6)]
wrong_cases = []
for i in range(NUMBER_OF_CASES):
    print("====================================================")
    print(f"Evaluating test case {i+1} of {NUMBER_OF_CASES} with the length of public key n = {len(public_key)}")
    cProfile.run("FindSubsetSum(set=public_key, n=len(public_key), sum=C)")
    public_key = extended + public_key
    extended = [0 for x in range(STEP)]
```
