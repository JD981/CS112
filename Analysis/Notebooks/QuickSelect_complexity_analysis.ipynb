# Quick Select time analysis

## Quick Select algorithm

```
def QuickSelect(left, right, a, k):
  if (left == right):
    return a[left]
 
  #index = Partition(a, left, right)
  if(left < right):
    index = left -1
    for j in range(left, right):
      if (a[j] >= a[right]):
        index += 1
        a[index], a[j] = a[j], a[index]
 
    a[index + 1], a[right] = a[right], a[index + 1]
    index += 1
 
    if (k == index):
      return a[k]
    if (k < index):
      return QuickSelect(left, index-1, a, k)
    return QuickSelect(index+1, right, a, k)  
def kth_element_qs(arr, k):
  return QuickSelect(0,len(arr)-1,arr, k-1)  
```

## Time calculating

```
import timeit
import random

for j in range(1, 40):
  a=[i for i in range(2**j)]
  random.shuffle(a)
  k = random.randint(1, len(a)-1)
  print("len = ", len(a), " k = ", k)
  %timeit kth_element_qs(a[:], k)
```

# Coefficient Calculating

## Initialization

```
import numpy as np
from sys import stdout
X = data['len(arr)']
Y = data['time']

X_logn = np.log2(X)
X_n = X
X_nlogn = X * np.log2(X)
X_n2 = pow(X, 2)
stdout.write("X_logn\n" + str(X_logn) + "\n\nX_n\n" + str(X_n) + \
             "\n\nX_nlogn\n" + str(X_nlogn) + "\n\nX_n2\n" + str(X_n2))
```

## O(log(n))

```
from sklearn.linear_model import LinearRegression

model = LinearRegression()
_X_logn = X_logn.values.reshape(-1, 1)
model.fit(_X_logn, Y)

print("{0:.20f}".format(model.coef_[0]))
#model.intercept_
```

## O(n)

```
from sklearn.linear_model import LinearRegression

model = LinearRegression()
_X_n = X_n.values.reshape(-1, 1)
model.fit(_X_n, Y)

print("{0:.20f}".format(model.coef_[0]))
#model.intercept_
```

## O(n.log(n))

```
from sklearn.linear_model import LinearRegression

model = LinearRegression()
_X_nlogn = X_nlogn.values.reshape(-1, 1)
model.fit(_X_nlogn, Y)

print("{0:.20f}".format(model.coef_[0]))
#model.intercept_
```

## O(n^2)

```
from sklearn.linear_model import LinearRegression

model = LinearRegression()
_X_n2 = X_n2.values.reshape(-1, 1)
model.fit(_X_n2, Y)

print("{0:.20f}".format(model.coef_[0]))
#model.intercept_
```
