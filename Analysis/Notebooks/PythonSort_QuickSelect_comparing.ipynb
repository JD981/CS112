```
#Quick select
def QuickSelect(left, right, arr, k):
  global count
  count += 1
  if (left == right):
    return arr[left]
 
  #index = Partition(a, left, right)
  if(left < right):
    index = left -1

    for j in range(left, right):
      count += 1
      if (arr[j] >= arr[right]):
        index += 1
        arr[index], arr[j] = arr[j], arr[index]
    arr[index + 1], arr[right] = arr[right], arr[index + 1]

    index += 1
    if (k == index):
      return arr[k]

    if (k < index):
      return QuickSelect(left, index-1, arr, k)
    return QuickSelect(index+1, right, arr, k)    
```

```
!cat QSelect.csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('QSelect.csv')
data.columns = ['size', 'operations', 'time']

X = data['size']
Y = data['operations']

X_nlogn = X * np.log2(X)
X_cann = np.sqrt(X)
X_logn = np.log2(X)
X_n2 = pow(X, 2)
X_n = X
```

```
import math as m
from sklearn.linear_model import LinearRegression

Y = data['operations'].values.reshape(-1, 1)

a_nlogn = LinearRegression()
a_cann = LinearRegression()
a_logn = LinearRegression()
a_n2 = LinearRegression()
a_n = LinearRegression()

nlogn = X_nlogn.values.reshape(-1, 1)
cann = X_cann.values.reshape(-1, 1)
logn = X_logn.values.reshape(-1, 1)
n2 = X_n2.values.reshape(-1, 1)
nn = X_n.values.reshape(-1, 1)


a_nlogn.fit(nlogn, Y)
a_cann.fit(cann, Y)
a_logn.fit(logn, Y)
a_n2.fit(n2, Y)
a_n.fit(nn, Y)

print(a_nlogn.coef_, ", ", a_nlogn.intercept_)
print(a_cann.coef_, ", ", a_cann.intercept_)
print(a_logn.coef_, ", ", a_logn.intercept_)
print(a_n2.coef_, ", ", a_n2.intercept_)
print(a_n.coef_, ", ", a_n.intercept_)
```

```
import math as m

fham = open("hamQS.txt", "w")
fham.write("nlogn,,cann,,logn,,n^2,n\n")

mse_nlogn = 0
mse_cann = 0
mse_logn = 0
mse_n2 = 0
mse_n = 0

for i in range(0, 100):
  k = int(nlogn[i]*a_nlogn.coef_ + a_nlogn.intercept_)
  fham.write(str(k) + "," + str(m.pow(Y[i] - k, 2)) + ",")
  mse_nlogn += m.pow(Y[i] - k, 2)

  k = int(cann[i]*a_cann.coef_ + a_cann.intercept_)
  fham.write(str(k) + "," + str(m.pow(Y[i] - k, 2)) + ",")
  mse_cann += m.pow(Y[i] - k, 2)

  k = int(logn[i]*a_logn.coef_ + a_logn.intercept_)
  fham.write(str(k) + "," + str(m.pow(Y[i] - k, 2)) + ",")
  mse_logn += m.pow(Y[i] - k, 2)

  k = int(n2[i]*a_n2.coef_ + a_n2.intercept_)
  fham.write(str(k) + "," + str(m.pow(Y[i] - k, 2)) + ",")
  mse_n2 += m.pow(Y[i] - k, 2)

  k = int(nn[i]*a_n.coef_ + a_n.intercept_)
  fham.write(str(k) + "," + str(m.pow(Y[i] - k, 2)) + "\n")
  mse_n  += m.pow(Y[i] - k, 2)

fham.write("," + str(mse_nlogn / 100) + ",," + str(mse_cann / 100) + ",," + str(mse_logn / 100) + ",," + str(mse_n2 / 100) + ",," + str(mse_n / 100))

fham.close()
```
